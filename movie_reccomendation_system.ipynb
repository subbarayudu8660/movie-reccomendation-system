```python
import numpy as np
import pandas as pd
import ast
```

```python
import kagglehub
path = kagglehub.dataset_download("tmdb/tmdb-movie-metadata")
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
credits.head()
```

```python
movies = movies.merge(credits,on = "title")
movies.head(1)
```

```python
movies = movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python

movies.drop_duplicates(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
  lis = []
  for i in ast.literal_eval(obj):
    lis.append(i["name"])
  return lis
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

When rows are removed from a DataFrame (e.g., duplicates or rows with missing values), the index labels of the remaining rows are preserved. This means that if rows with indices 5, 10, and 15 were removed, those index values would no longer exist in the DataFrame, but the subsequent indices would not shift up. As a result, the maximum index value might be higher than the actual number of rows.

To reset the index to a sequential range starting from 0, you can use the `reset_index()` method.

```python
movies.reset_index(drop=True, inplace=True)
movies.head(1)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
def convert3(obj):
  lis = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!=3:
      lis.append(i["name"])
      counter+=1
    else:
      break
  return lis
```

```python
movies["cast"] = movies["cast"].apply(convert3)
```

```python
movies["crew"][0]
```

```python
def convert_crew(obj):
  lis = []
  for i in ast.literal_eval(obj):
    if i["job"] == "Director":
      lis.append(i["name"])
      break
  return lis
```

```python
movies["crew"] = movies["crew"].apply(convert_crew)
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```
